```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, TensorBoard
from tensorflow.keras.optimizers import SGD,Adamax,RMSprop
import numpy as np
from sklearn.utils import shuffle
from sklearn.metrics import confusion_matrix,f1_score
import difflib
import time
import matplotlib.pyplot as plt
from modulos.machine_learning import carrega_dados, kfold
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split
#supress notation
np.set_printoptions(suppress=True)
```

```python
x, y, x_teste, y_teste = carrega_dados()
```

```python
[x.shape,y.shape],[x_teste.shape,y_teste.shape]
```

```python
#optionally, join test set with train set.
#x = np.concatenate((x,x_teste),axis=0)
#y = np.concatenate((y,y_teste),axis=0)
#x.shape,y.shape
```

```python
model = Sequential()
model.add(Dense(units=73, activation='relu', kernel_initializer='random_uniform', input_dim = 73))
model.add(Dropout(0.4))
model.add(Dense(units = 128, activation = 'relu'))

model.add(Dense(units = 1, activation = 'sigmoid'))
otimizador = SGD(momentum=0.9, nesterov=True, learning_rate=0.005)
model.compile(optimizer = otimizador, loss = 'binary_crossentropy',
                  metrics = ['accuracy'])
```

```python
model.summary()
```

```python
x, y = shuffle(x , y)
kf = KFold(n_splits=3)
#kf = StratifiedKFold(n_splits=5, shuffle=True)
cvscores = []
histories = []
for treino, teste in kf.split(x, y):
    history = model.fit(x[treino], y[treino], epochs=200, batch_size=20, verbose=0)
    scores = model.evaluate(x_teste, y_teste, verbose=0)
    cvscores.append(scores[1] * 100)
    histories.append(history)
    print(scores)
print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores))) # Resultado Final
```
